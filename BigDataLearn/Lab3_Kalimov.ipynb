```python
import math
import numpy as np
import matplotlib.pyplot as plt
# Импортируем пакет со вспомогательными функциями
import matplotlib.mlab as ml
```

```python
# 2

%matplotlib inline
x = np.linspace(-8,10,100)
y = 2 * (x - 1)**2 + 4
plt.plot(x,y, label=' y = 2 * (x - 1)^2 + 4')
plt.legend(loc='best')
plt.title('График квадратичной функции')
plt.xlabel('Значение по оси X')
plt.ylabel('Значение по оси Y')
plt.show()
```

```python
#3
%matplotlib inline
x = np.linspace(-5*np.pi,5*np.pi,1000)
y = np.sin(x)
plt.plot(x,y, label='16 + np.sin(3*x) + 1')
plt.legend(loc='best')
plt.title('График гармонической функции')
plt.grid(color = 'w', linewidth = 1)
plt.xlabel('Значение по оси X')
plt.ylabel('Значение по оси Y')
plt.grid(True)
plt.show()
```

```python
# 4
fig = plt.figure(figsize=(8, 3))

x = np.arange(0, 10, 0.1)
y = np.cos(x*2)
z = np.exp(-x) 
ax = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax.set_title('Косинусоида')
ax.plot(y, color='g', label='Косинусоида')

ax2.set_title('Экспонента')
ax2.plot(z, color='r', label='Экспонента')

plt.show()
```

```python
# 5
fig = plt.figure(figsize=(8, 3))

x = np.arange(0, 10, 0.1)
y = np.cos(x)
z = np.exp(-x) 
ax = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax.set_title('Общая картина')
ax.plot(y, color='g', label='Косинусоида')
ax.plot(z, color='r', label='Экспонента')

ax2.set_title('Усеченная версия')
ax2.plot(y, color='g', label='Косинусоида')
ax2.plot(z, color='r', label='Экспонента')
ax2.set_xlim([15, 35])

plt.show()
```

```python
# 6 
q = np.linspace(0, 0.4*np.pi, 1000)
x = np.sin(q*20)
y = np.cos(q*50)
plt.figure(figsize=(7, 7))
plt.plot(x, y, label='Параметрическая функция cos(q), sin(q)')
plt.title('График параметрической функции')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend(loc='best')
plt.show()
```

```python
# 7
subsistence_level = [4794.34, 4900.79,4792.54, 4806.32,5358.58, 6268.58, 7178.32]
years = [2016, 2017, 2018, 2019, 2020, 2021, 2022]
dpi = 80

plt.figure(dpi = dpi, figsize = (512 / dpi, 384 / dpi) )
plt.bar(years, subsistence_level, color='g')
plt.title('Прожиточный минимум в Кыргызстане по годам')
plt.xlabel('Год')
plt.ylabel('Прожиточный минимум в сомах')
plt.grid(axis='y')
plt.show()
```

```python
# 8
dpi = 80 
fig = plt.figure(dpi = dpi, figsize = (512 / dpi, 384 / dpi) )
# mpl.rcParams.update({'font.size': 8})
plt.rcParams.update({'font.size': 8})
plt.title('Распределение по годам')
xs = range(len(years))
plt.pie(
    subsistence_level, autopct='%.1f', radius = 1.1,
    explode = [0.15] + [0 for _ in range(len(years) - 1)] )
plt.legend(
    bbox_to_anchor = (-0.16, 0.45, 0.25, 0.25),
    loc = 'lower left', labels = years )
plt.show()
```

```python
# 9
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)

u, v = np.meshgrid(u, v)
a = 1.0
b = 0.5

x = (a + b * np.cos(v)) * np.cos(u)
y = (a + b * np.cos(v)) * np.sin(u)
z = b * np.sin(v)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, color='b')
plt.show()
```

```python
u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
x = np.cos(u)*np.sin(v)
y = np.sin(u)*np.sin(v)
z = np.cos(v)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(x, y, z)
ax.legend()
plt.show()
```
